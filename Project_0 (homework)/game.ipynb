# Игра "Угадай число"

## Версия 1

### Система загадывает число, которое пользователь должен угадать, вводя ответ с клавиатуры

```python
import numpy as np

count = 0

number = np.random.randint(1,101)

while True:
    predict_number = int(input("Введите число: "))
    count+=1
    if predict_number > number:
        print("Введенное число больше искомого \n")
    elif predict_number < number:
        print("Введенное число меньше искомого \n")
    else:
        print(f"Поздравляем! Вы угадали число {number} за {count} попыток! \n")
        break
```
